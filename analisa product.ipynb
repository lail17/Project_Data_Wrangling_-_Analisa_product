## 1. 10 produk yang paling banyak diminati

### Dataset yang digunakan

1. df_products

2. df_product_translation

3. df_items

### Mengolah Dataset

```python
df_minat = pd.read_sql_query(
    "with minat as(select product_category_name, count(product_id) as total_order \
    from olist_products_dataset inner join olist_order_items_dataset using(product_id) \
    group by product_category_name) \
    select * from minat order by total_order desc limit 10"
    , con)
df_minat
```

```python
df_minat_english=pd.merge(df_minat,df_product_translation, on='product_category_name',how='left')
df_minat_baru = df_minat_english[['product_category_name_english','total_order']]
df_minat_baru
```

### cek missing data

```python
df_minat_baru.isna().sum()
```

### cek duplicate data

```python
df_minat_baru.duplicated().sum()
```

### Cek inconsistent format

```python
df_minat_baru['product_category_name_english'].unique()
```

### Visualisasi

```python
df_minat_sorted = df_minat_baru.sort_values('total_order')
df_minat_sorted.plot.barh(x ='product_category_name_english', color = "magenta")
```

## 2. 10 kota dengan pendapatan tertinggi

### Dataset yang digunakan

1. df_sellers

2. df_items

### Mengolah Dataset

```python
df_pendapatan_kota = pd.read_sql_query(
    "with pendapatan_kota as(select seller_city, sum(price) as total_pendapatan \
    from olist_sellers_dataset inner join olist_order_items_dataset using(seller_id) \
    group by seller_city) \
    select * from pendapatan_kota order by total_pendapatan desc limit 10"
    , con)
df_pendapatan_kota
```

### cek missing data

```python
df_pendapatan_kota.isna().sum()
```

### cek duplicate data

```python
df_pendapatan_kota.duplicated().sum()
```

### Cek inconsistent format

```python
df_pendapatan_kota['seller_city'].unique()
```

### Visualisasi

```python
df_pendapatan_kota_sorted = df_pendapatan_kota.sort_values('total_pendapatan')
df_pendapatan_kota_sorted.plot.barh(x ='seller_city', color = "magenta")
```

## 2. 10 kota dengan jumlah pembeli terbanyak

### Dataset yang digunakan

1. df_customer

2. df_order

### Mengolah Dataset

```python
df_pembelian_kota = pd.read_sql_query(
    "with pembelian_kota as(select customer_city, count(order_id) as total_pembelian \
    from olist_order_customer_dataset inner join olist_order_dataset using(customer_id) \
    group by customer_city) \
    select * from pembelian_kota order by total_pembelian desc limit 10"
    , con)
df_pembelian_kota
```

### cek missing data

```python
df_pembelian_kota.isna().sum()
```

### cek duplicate data

### Cek inconsistent format

### Visualisasi

```python
df_pembelian_kota_sorted = df_pembelian_kota.sort_values('total_pembelian')
df_pembelian_kota_sorted.plot.barh(x ='customer_city', color = "magenta")
```
